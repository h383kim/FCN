```python
import pickle as pkl
import matplotlib.pyplot as plt

with open('loss_acc_dict.pkl', 'rb') as f:
    loss_acc_dict = pkl.load(f)

epochs = range(1, 61)
# Train Loss Graph
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(epochs, loss_acc_dict['train_loss_lst'], label="Train Loss", linestyle="-")
plt.plot(epochs, loss_acc_dict['val_loss_lst'], label="Validation Loss", linestyle="--")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Train / Val Loss Graph")
plt.legend()

# Train Accuracy Graph
plt.subplot(1, 2, 2)
plt.plot(epochs, loss_acc_dict['train_acc_lst'], label="Train Accuracy", linestyle="-")
plt.plot(epochs, loss_acc_dict['val_acc_lst'], label="Validation Accuracy", linestyle="--")
plt.xlabel("Epoch")
plt.ylabel("Accuracy (%)")
plt.title("Train / Val Accuracy Graph")
plt.legend()
```
